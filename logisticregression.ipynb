```python
# predicting if a person would buy life insurance based on his age using logistic regression
```

```python
import pandas as pd
from matplotlib import pyplot as plt
%matplotlib inline
```

```python
 df = pd.read_csv("insurance_data.csv")
df.head()
```

```python
plt.scatter(df.age,df.bought_insurance,marker='+',color='purple')
```

```python
df.shape
```

```python
# first split the dataset into training and testing data
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(df[['age']],df.bought_insurance,train_size=0.8) #80% for training
```

```python
X_test
```

```python
X_train
```

```python
from sklearn.linear_model import LogisticRegression
model = LogisticRegression()
```

```python
model.fit(X_train, y_train) # train
```

```python
model.predict(X_test)
```

```python
model.score(X_test, y_test)
```

```python
# predict on test data
y_pred = model.predict(X_test)

for i in range(len(X_test)):
    print(f"Predicted: {y_pred[i]}, Actual: {y_test.values[i]}")
```

```python
# predicting if a person aged 30 would buy insurance
new_person = [[30]]
prediction = model.predict(new_person)
print(f"Prediction for age 30: {prediction[0]}")  
```
